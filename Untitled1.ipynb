```python
data = df[['City', 'AQI', 'PM2.5', 'PM10']].dropna().to_numpy()


cities = np.unique(data[:, 0])


result = {}
for city in cities:
    city_data = data[data[:, 0] == city][:, 1:].astype(float) 
    result[city] = {
        'Mean AQI': np.mean(city_data[:, 0]),
        'Mean PM2.5': np.mean(city_data[:, 1]),
        'Mean PM10': np.mean(city_data[:, 2])
    }


for city, stats in result.items():
    print(f"City: {city}")
    print(f"  Mean AQI: {stats['Mean AQI']}")
    print(f"  Mean PM2.5: {stats['Mean PM2.5']}")
    print(f"  Mean PM10: {stats['Mean PM10']}")
```

```python
# check for missing values
df.isnull()
```

```python
# replace missing values with column mean and print the update DataFrame
numeric_columns = df.select_dtypes(include=["number"]).columns
for column in numeric_columns:
    df[column].fillna(df[column].mean())
print(df)
```
